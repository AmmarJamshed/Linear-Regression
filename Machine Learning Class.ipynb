```python
import pandas as pd
import numpy as np
import sklearn
```

```python
dfh = pd.read_excel(r'C:\Users\muham\Downloads\Hour Price Prediction.csv.xlsx')
```

```python
dfh.head()
```

```python
dfh.drop(['Id'],axis=1, inplace=True) # we use in place to drop the whole column
```

```python
x = dfh.drop(['SalePrice'],axis=1)
```

```python
y = dfh['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
# Seperate test and training data
Tx, Tex, Ty, Tey = train_test_split(x, y, test_size=0.4, random_state=0)
```

```python
# import linear Regression 
from sklearn.linear_model import LinearRegression
```

```python
ModelLR = LinearRegression()
```

```python
ModelLR = ModelLR.fit(Tx, Ty)
```

```python
ModelLR.intercept_
```

```python
ModelLR.coef_
```

```python
# Predict with test data only
ModelLR.predict(Tex)
```

```python
prediction = ModelLR.predict(Tex)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2_score(prediction, Tey)
```

```python
mean_absolute_error(prediction, Tey)
```

```python
mean_squared_error(prediction, Tey)
```
